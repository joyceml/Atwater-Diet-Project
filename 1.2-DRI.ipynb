```python
#!pip install -r requirements.txt
#pip install --upgrade oauth2client
#!pip install gspread
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import pandas as pd
```

```python
scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']
json_file = 'southern-ivy-270522-0f3304361442.json' #UNIQUE JSON FILE FOR INDIVIDUAL'S GOOGLE API CREDENTIALS
spreadsheet = '1oz_3tKqIaIrCd-OeAVcBfEoC83WJtstQeKWLQDt2ZlU' #URL FOR DRI GOOGLE SHEETS
credentials = ServiceAccountCredentials.from_json_keyfile_name(json_file, scope)


gc = gspread.authorize(credentials) #USE GOOGLE API CREDENTIALS TO ACCESS DRI VALUES FROM GOOGLE SHEETS
sh = gc.open_by_key(spreadsheet)
wks = sh.get_worksheet(0)
wks = sh.sheet1 #TURN GOOGLE SHEET INTO WORKSHEET
data = wks.get_all_values()
headers = data.pop(0)
data_df = pd.DataFrame(data,columns = headers) #TURN WORKSHEET TO A PANDAS DATAFRAME
data_df
```

```python
def daily_recommended_intake(age, gender):
    if age <= 0: #FILTERS DATAFRAME BY AGE, THEN GENDER
        return 'Error: Age must be positive'
    else: 
        if age <= 3: 
            print('Recommended Values for Females and Males')
            return data_df['Female and Male (Ages 1-3)'] #RETURNS COLUMN WITH DESIRED STRING AS TITLE
        if age >= 4 and age <= 8:
            if gender == "female":
                return data_df['Female (Ages 4-8)']
            if gender == "male":
                return data_df['Male (Ages 4-8)']
        if age >= 9 and age <= 13:
            if gender == "female":
                return data_df['Female (Ages 9-13)']
            if gender == "male":
                return data_df['Male (Ages 9-13)'] 
        if age >= 14 and age <= 18:
            if gender == "female":
                return data_df['Female (Ages 14-18)']
            if gender == "male":
                return data_df['Male (Ages 14-18)'] 
        if age >= 19 and age <= 30:
            if gender == "female":
                return data_df['Female (Ages 19-30)']
            if gender == "male":
                return data_df['Male (Ages 19-30)'] 
        if age >= 31 and age <= 50:
            if gender == "female":
                return data_df['Female (Ages 31-50)']
            if gender == "male":
                return data_df['Male (Ages 31-50)'] 
        if age >= 51:
            if gender == "female":
                return data_df['Female (Ages 51+)']
            if gender == "male":
                return data_df['Male (Ages 51+)'] 
            
    
```
